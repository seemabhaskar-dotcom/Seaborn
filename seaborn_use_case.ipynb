```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
#load the data set
tips=sns.load_dataset("tips")
```

```python
sns.pairplot(tips)
plt.show()
```

```python
sns.lineplot(x="total_bill", y="tip", data=tips)
plt.show()
```

```python
tips.columns
```

```python
tips["sex"].value_counts().plot.pie(autopct="%1.1f%%", colors=["grey", "purple"])
plt.title("male and female in percentage")
plt.legend(loc="best")
plt.show()
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="time")
plt.title("Data visualization of tips")
plt.show()
```

```python
tips.plot.scatter(x="total_bill", y="tip")
```

```python
sns.histplot(data=tips, x="total_bill", bins=20, kde= True)   #KDE Plot described as Kernel Density Estimate is used for visualizing the Probability Density 
                                                              #of a continuous variable. It depicts the probability density at different values in a continuous variable.
plt.xlabel("total_bill($)")
plt.ylabel("count")
plt.show()
```

```python
#Normal distribution

from scipy.stats import norm
x=np.linspace(-4,4,1000)
plt.plot(x, norm.pdf(x, loc=0, scale=1), "r-", label="Normal Distribution")
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.title('Normal Distribution Curve')
plt.legend()
plt.grid()
plt.show()
```
